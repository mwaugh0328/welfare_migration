```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec
import scipy.io
import seaborn as sns
```

```python
low_mat = scipy.io.loadmat( "low_z_move.mat")

low_move = pd.DataFrame(low_mat["low_z"])

med_mat = scipy.io.loadmat( "med_z_move.mat")

med_move = pd.DataFrame(med_mat["med_z"])

exp_mat = scipy.io.loadmat( "exp_z_move.mat")

exp_move = pd.DataFrame(exp_mat["exp_z"])
```

### Migration Decision

```python
fig, ax = plt.subplots(1,1,figsize = (16,9))

#cbar_ax = fig.add_axes([0.95, .25, .03, .5])

sns.heatmap((1-low_move).iloc[50:100,0:30], cmap="Blues",cbar = True,vmin=0,vmax=1.0, ax=ax,
           cbar_kws=dict(use_gridspec=False,location="right",label = "\n Probability of Moving"))

ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

ax.text(-2, 50, "Larger Asset Holdings ⟶ ",fontsize = 14, rotation = 90)

ax.text(0, 53, "Larger Transitory Shock ⟶ ",fontsize = 14)

ax.set_title("Migration Policy Function: Low z, No Experinced \n", fontsize = 16)

fig_path = "C:\\github\\welfare_migration\\slides\\TALK-FIGURES\\"

plt.savefig(fig_path + "\\migration_policy_low_z.pdf", bbox_inches = "tight", dip = 3600)

plt.show()
```
